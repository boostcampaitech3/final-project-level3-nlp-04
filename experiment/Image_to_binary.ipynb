# Image to Binary
- Api에 image 전송 시 file path, url, binary object 전송 가능
- cv2를 사용하여 가공된 image를 binary화 하여 api 전송 가능
- binary로 변환 과정 중 손실이 있는지 확인 필요

```python
import cv2
file_path = f"/opt/ml/img/raw_image_1.jpg"
img = cv2.imread(file_path)
print("image type : ", type(img))

binary_cv = cv2.imencode('.PNG', img)[1].tobytes()
print("binary 변환 : ", type(binary_cv))
```

```python
import requests
api_url = "http://118.222.179.32:30001/ocr/"
headers = {"secret": "Boostcamp0001"}
file_dict = {"file": binary_cv}
response = requests.post(api_url, headers=headers, files=file_dict)
ocr_output = response.json()
ocr_output
```
