### 문제 정의 : 거리가 가까운 단어가 띄워져서 각각을 따로보는 현상을 해결하는 코드를 만들어보자

### WorkFlow
    1. point를 참고하여 어떤 부분이 width인지, length인지 파악
    2. 한 글자의 평균 width를 구해서 평균 width 차이가 나지 않는 그룹을 병합시킴

```python
file_path = "/opt/ml/final/test_image/test1_101rotate.jpg"

# Python
import requests
api_url = "http://118.222.179.32:30001/ocr/"
headers = {"secret": "Boostcamp0001"}
file_dict = {"file": open(file_path, "rb")}
response = requests.post(api_url, headers=headers, files=file_dict)
ocr_output = response.json();

test_list = ocr_output['ocr']['word']
# test_list.sort(key = lambda x : x['points'][0])

for i in range(len(test_list)) : 
    print(test_list[i])
```

```python
# 데이터프레임 만들어보자
import pandas as pd
point_1_x = []
point_1_y = []
point_2_x = []
point_2_y = []
point_3_x = []
point_3_y = []
point_4_x = []
point_4_y = []
ocr_text = []
word_length = []
box_width = []
box_length = []
english = []

# 일단 모두다 horizontal인 경우로 가정한다.
for i in range(len(test_list)) :
    point_1_x.append(test_list[i]['points'][0][0])
    point_1_y.append(test_list[i]['points'][0][1])
    point_2_x.append(test_list[i]['points'][1][0])
    point_2_y.append(test_list[i]['points'][1][1])
    point_3_x.append(test_list[i]['points'][2][0])
    point_3_y.append(test_list[i]['points'][2][1])
    point_4_x.append(test_list[i]['points'][3][0])
    point_4_y.append(test_list[i]['points'][3][1])
    # 1 ~ 2 간의 거리
    temp_length1 = (abs(test_list[i]['points'][1][0] - test_list[i]['points'][0][0])**2 + abs(test_list[i]['points'][1][1] - test_list[i]['points'][0][1])**2)**0.5
    # 2 ~ 3 간의 거리
    temp_length2 = (abs(test_list[i]['points'][2][0] - test_list[i]['points'][1][0])**2 + abs(test_list[i]['points'][2][1] - test_list[i]['points'][1][1])**2)**0.5
    
    ocr_text.append(test_list[i]['text'])
    word_length.append(len(test_list[i]['text']))
    box_width.append(max(temp_length1, temp_length2))
    box_length.append(min(temp_length1, temp_length2))
    english.append(test_list[i]['text'].encode().isalpha())

ocr_df = pd.DataFrame({'text' : ocr_text,
                       'is_alpha' : english,
                       'word_length' : word_length,
                       'box_width' : box_width,
                       'box_length' : box_length,
                       'point_1_x' : point_1_x,
                       'point_1_y' : point_1_y,
                       'point_2_x' : point_2_x,
                       'point_2_y' : point_2_y,
                       'point_3_x' : point_3_x,
                       'point_3_y' : point_3_y,
                       'point_4_x' : point_4_x,
                       'point_4_y' : point_4_x,})
ocr_df
```

```python
# 한글, 영어 여부에 따라 한글자의 width가 달라짐
# 글자 크기에 따라서도 달라지니깐 이 부분도 해결해야할 듯
eng_cnt = 0
eng_length = 0

kor_cnt = 0
kor_length = 0

for i in range(len(ocr_df)) :
    # 영어면 영어끼리 길이 구함
    if ocr_df['is_alpha'][i] == True : 
        eng_cnt += ocr_df['word_length'][i]
        eng_length += ocr_df['box_width'][i]
    # 한국어면 한국어끼리 길이 구함
    else : 
        kor_cnt += ocr_df['word_length'][i]
        kor_length += ocr_df['box_width'][i]
        
print(f'영어 한글자 width 평균 = {eng_length / eng_cnt}')
print(f'한국어 한글자 width 평균 = {kor_length / kor_cnt}')
```
